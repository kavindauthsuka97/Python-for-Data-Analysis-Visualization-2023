#### figure ()

```python
fig = plt.figure() #creating the figure object
```

```python
# add a rectangular
fig = plt.figure().add_axes(rect=[0,0,1,0.5])
#third parameter = width
#fourth parameter = height
```

```python
# we can do above thing like this too
fig = plt.figure(figsize=(5,5)) #object
axes = fig.add_axes(rect=[0,0,1,0.5]) #area of the figure
# first element = position from the left
# second element =  position from the bottom
# third element = width
# fourth element = height
```

#### Adjusting the axis

```python
fig = plt.figure(figsize=(5,5))
axes1 = fig.add_axes(rect=[0,0,0.5,0.5])
axes2 = fig.add_axes(rect=[0,1,0.5,0.5])
axes1.set_ylim(0,100) #change the Y axis 
axes2.set_xlim(0,50) #change the X axis
```

#### twinx()
* Add two axes on the same chart

```python
fig = plt.figure(figsize=(10,10))
axes1 = fig.add_axes(rect=[0,0,0.5,0.5])
axes2 = axes1.twinx()

axes1.set_ylim(0,100)
```

#### Add chart title

```python
fig = plt.figure(figsize=(5,5))
axes1 = fig.add_axes(rect=[0,0,0.5,0.5])
axes2 = fig.add_axes(rect=[1,0,0.5,0.5])

#add title to the graph
axes1.set_title('Chart 1')
axes2.set_title('Chart 2')

#add labels
axes1.set_xlabel('X')
axes1.set_ylabel('Y')

axes2.set_xlabel('X')
axes2.set_ylabel('Y')
```

## Bar Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('top_10_countries.csv')
```

```python
df.head()
```

```python
df.rename(columns={'Country / Dependency':'Country'},inplace=True)
```

```python
df.head()
```

```python
df_g = df[['Region','Population']]
```

```python
df_g = df_g.groupby(by='Region').sum()
```

```python
df_g.reset_index(inplace=True)
```

```python
fig = plt.figure(figsize=(5,5)) # creating figure object
axes = fig.add_axes([0,0,1,1]) #create an area
axes.set_xlabel('Region')
axes.set_ylabel('Total Population')
axes.set_title('Top 10 Countries Population by Region')

#bar plot
plt.bar(x=df_g['Region'],height=df_g['Population'],color = '#5584AC')
# edgecolor ->  we can pass the color for edges using this parameter
plt.show()
```

```python
#bar plot
# we can pass different colors to each bar
plt.bar(x=df_g['Region'],height=df_g['Population'],color = ['b','r','y','g'])
# edgecolor ->  we can pass the color for edges using this parameter
plt.show()
```

## Horizontal Bar Plots

```python
fig = plt.figure(figsize = (5,5))
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([1.5,0,1,1])
```

```python
# we can do the above thing using subplots

#creating figure and axes
fig, [ax1,ax2] = plt.subplots(nrows=1,ncols=2,figsize=(20,8))

#labels and titles
ax1.set_title('Total Population by Region')
ax1.set_ylabel('Population')
ax1.set_xlabel('Region')

ax2.set_title('Total Population by Country')
ax2.set_ylabel('Population')
ax2.set_xlabel('Country')

# add bar plot
ax1.bar(x=df_g['Region'], height = df_g['Population'])

# horizontal bar
ax2.barh(y=df['Country'], width = df['Population'])
```

## Line Plots

```python
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt
```

```python
tfl = pd.read_csv('tfl-daily-cycle-hires.csv')
```

```python
tfl.drop(columns='Unnamed: 2', inplace=True)
```

```python
tfl['Day'] = pd.to_datetime(tfl['Day'],format='%d/%m/%Y')
```

```python
tfl.dtypes
```

```python
tfl['Day'] = tfl['Day'].astype('datetime64')
```

```python
tfl.sort_values('Day', inplace=True)
```

```python
tfl['Year'] = tfl['Day'].dt.strftime('%Y')
```

```python
tfl_y = tfl[['Year', 'Number of Bicycle Hires']]
```

```python
tfl_y
```

```python
tfl_y = tfl_y.groupby(by='Year').sum().reset_index()
```

```python
tfl_y
```

```python
# creating figure object

fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])

ax1.plot(tfl_y['Year'],tfl_y['Number of Bicycle Hires'],color='r',alpha=1,marker='o')
#there is no keyword arguments. there is only positional arguments
# alpha = brightness
# marker = we can add a marker to each point
```

## Histograms

```python
players = pd.read_csv('players_21.csv')
```

```python
# Creating a Hist Plot
fig = plt.figure(figsize = (5,5))

ax1 = fig.add_axes([0,0,1,1])
ax1.set_ylabel('Number of Players')
ax1.set_xlabel('Overall score (bins)')
plt.xticks([0,25,50,75,100],[0,25,50,75,100]) #add xticks for X axis
ax1.hist(x=players['overall'],bins = [0,25,50,75,100],edgecolor = 'black')

plt.show()
```

## Horizontal Histogram


```python
# Creating a Hist Plot
fig = plt.figure(figsize = (5,5))

ax1 = fig.add_axes([0,0,1,1])
ax1.set_ylabel('Number of Players')
ax1.set_xlabel('Overall score (bins)')
plt.xticks([0,25,50,75,100],[0,25,50,75,100]) #add xticks for X axis
ax1.hist(x=players['overall'],bins = [0,25,50,75,100],edgecolor = 'black',orientation='horizontal')
```

## Box Plots

```python
fig = plt.figure(figsize = (5,5))

ax1 = fig.add_axes([0,0,1,1])
ax1.boxplot(players['overall'],labels=['overall score'])
plt.show()
```

```python
fig = plt.figure(figsize = (5,5))

ax1 = fig.add_axes([0,0,1,1])
ax1.boxplot(players['overall'],labels=['overall score'],showfliers=False) 
#showfliers = False -> remove the outliers
plt.show()
```

## Multiple Box Plots

```python
# multiple boxplots
inter = players[players['club_name']=='Inter']['overall'] #filter the overall of 'inter' club
dortmund = players[players['club_name']=='Borussia Dortmund']['overall'] #filter the overall of 'Borussia Dortmund' club
data = [inter, dortmund] #create a list
data
```

```python
fig = plt.figure(figsize = (5,5))
ax1 = fig.add_axes([0,0,1,1])
ax1.boxplot(data,labels = ['Inter','Dortmund']) 
plt.plot()
```

## Violin Plots

```python
fig = plt.figure(figsize = (5,5))
ax1 = fig.add_axes([0,0,1,1])
ax1.violinplot(data)
plt.xticks([1,2],['Inter','Dortmund'])
# data is in list format
```

#### Box Plots and Violin Plots in the same Plot

```python
fig = plt.figure(figsize = (5,5))
ax1 = fig.add_axes([0,0,1,1])
ax1.violinplot(data)
ax1.boxplot(data)
plt.xticks([1,2],['Inter','Dortmund'])
```

## Style and Presentation

```python
import numpy as np
x = np.arange(0,10,0.1)
```

```python
x.shape
```

```python
y1 = x
y2 = x**2
y3 = x**3
```

```python
(y1.shape,y2.shape,y3.shape)
```

#### Resources
* https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_title.html
* https://matplotlib.org/stable/gallery/pyplots/pyplot_simple.html
* https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xticks.html
* https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.yticks.html

```python
fig = plt.figure(figsize = (6,6))
ax1 = fig.add_axes([0,0,1,1])

ax1.plot(x,y1)
ax1.plot(x,y2)
ax1.plot(x,y3)

ax1.set_xlim(0,10)
ax1.set_ylim(0,100)

ax1.set_title('Chart Title')
ax1.set_xlabel('x label')
ax1.set_ylabel('y label')

plt.show()
```

```python
# adding the title in left and right and middle of the graph
fig = plt.figure(figsize = (6,6))
ax1 = fig.add_axes([0,0,1,1])

ax1.plot(x,y1)
ax1.plot(x,y2)
ax1.plot(x,y3)

ax1.set_title('Chart Title Left', loc='left')
ax1.set_title('Chart Title', loc='center')
ax1.set_title('Chart Title Right', loc='right')
ax1.set_xlabel('x label', )
ax1.set_ylabel('y label')

plt.show()
```

#### Rotate the x axis direction

```python
fig = plt.figure(figsize = (6,6))
ax1 = fig.add_axes([0,0,1,1])

ax1.plot(x,y1)
ax1.plot(x,y2)
ax1.plot(x,y3)

ax1.set_title('Chart Title', fontsize=40, color='blue', fontname='Arial')
ax1.set_xlabel('x label', fontsize=20, fontname='Times New Roman', rotation='vertical') #rotate the x lable name
ax1.set_ylabel('y label', fontsize=20, fontname='Helvetica')

plt.show()
```
